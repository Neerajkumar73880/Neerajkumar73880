**Linear Regression from Scratch**

Importing Python Libraries

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

Generate Random Data Points

```python
import numpy as np


x = np.random.randint(1, 10, 20)
y = np.random.randint(100, 500, 20)

print("Generated x:", x)
print("Generated y:", y)
```

Fit a Linear Regression Model

```python

x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
slope = numerator / denominator

b = y_mean - (slope * x_mean)

print("Slope (w):", slop)
print("Intercept (b):", b)

y_pred = slop * x + b
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.show()
```

Implement the Cost Function

```python
def compute_cost(x, y, w, b):

    m = len(y)
    total = 0
    for i in range(m):
      y_pred = w * x[i] + b
      total += (y_pred - y[i]) ** 2
    return total / m



```

```python
values = np.arange(100, 1001)
costs = []
for w in values:
  cost = compute_cost(x, y, w, b)
  costs.append(cost)


plt.plot(values, costs)
plt.xlabel("Weight (w)")
plt.ylabel("Cost")
plt.title("Cost Function")
plt.show()
```

```python

import matplotlib.pyplot as plt

def cost_function(x, y, w, b):
  n = len(x)
  total_cost = 0
  for i in range(n):
    y_pred = w * x[i] + b
    total_cost += (y_pred - y[i]) ** 2
  return total_cost / (2 * n)

def find_best_fit_weight(x, y, b):
  min_cost = float('inf')
  best_w = None
  for w in range(100, 1001):
    cost = cost_function(x, y, w, b)
    if cost < min_cost:
      min_cost = cost
      best_w = w
  return best_w, min_cost

b = 1

best_w, min_cost = find_best_fit_weight(x, y, b)
print(f"Best-fit weight (w): {best_w}")
print(f"Minimum cost: {min_cost}")



plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot of Data Points")
plt.show()

```
