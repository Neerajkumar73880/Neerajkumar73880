**Linear Regression from Scratch**

Importing Python Libraries

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Generate random points**

```python
import numpy as np


x = np.random.randint(10, 50, 20)
y = np.random.randint(100, 1000, 20)

print("Generated x:", x)
print("Generated y:", y)
```

Fit a Linear Regression Model

```python

x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
slope = numerator / denominator

b = y_mean - (slope * x_mean)

print("Slope (w):", slope)
print("Intercept (b):", b)

y_pred = slope * x + b
plt.scatter(x,y)
plt.plot(x, y_pred, color='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.show()
```

Implement the Cost Function

```python
def compute_cost(x, y, w, b):

    m = len(y)
    total = 0
    for i in range(m):
      y_pred = w * x[i] + b
      total += (y_pred - y[i]) ** 2
    return total / m


values = np.arange(100, 1001)
costs = []
for w in values:
  cost = compute_cost(x, y, w, b)
  costs.append(cost)


plt.plot(values, costs)
plt.xlabel("Weight (w)")
plt.ylabel("Cost")
plt.title("Cost Function")
plt.show()
```

**Optimization Function**

```python
import matplotlib.pyplot as plt

def cost_function(x, y, w, b):
  n = len(x)
  total_cost = 0
  for i in range(n):
    y_pred = w * x[i] + b
    total_cost += (y_pred - y[i]) ** 2
  return total_cost / (2 * n)

def find_best_fit_weight(x, y, b):
  min_cost = float('inf')
  best_w = None
  for w in range(100, 1001):
    cost = cost_function(x, y, w, b)
    if cost < min_cost:
      min_cost = cost
      best_w = w
  return best_w, min_cost

b = 1

best_w, min_cost = find_best_fit_weight(x, y, b)
print(f"Best-fit weight (w): {best_w}")
print(f"Minimum cost: {min_cost}")

```

**Visualize the original random data points as a scatter plot.**

```python

import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot of Data Points")
plt.show()
```

**The best-fit line obtained using your Linear Regression model**

```python

learning_rate = 0.0001
num_iterations = 100

w = 0
b = 0
for i in range(num_iterations):
    y_pred = w * x + b

    dw = -(2 / len(x)) * np.sum(x * (y - y_pred))
    db = -(2 / len(x)) * np.sum(y - y_pred)

    w = w - learning_rate * dw
    b = b - learning_rate * db

    plt.scatter(x, y)
    plt.plot(x, y_pred, color='red')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title(f"Linear Regression - Iteration {i+1}")
    plt.show()
    plt.pause(0.1)
    plt.clf()

print(f"Final weight (w): {w}")
print(f"Final bias (b): {b}")

```

**cost function J(w) for different values of w.**

```python

w_values = np.linspace(-10, 10, 100)

costs = [compute_cost(x, y, w, b) for w in w_values]

# Plot the cost function
plt.plot(w_values, costs)
plt.xlabel("Weight (w)")
plt.ylabel("Cost (J(w))")
plt.title("Cost Function J(w)")
plt.show()
```
